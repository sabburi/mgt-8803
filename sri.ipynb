```python
import pandas as pd
import numpy as np 
import sqlite3

conn = sqlite3.connect("8803.db")
```

```python
edges = pd.read_csv('csv_paradise_papers.2018-02-14/paradise_papers.edges.csv', low_memory = False)
address = pd.read_csv('csv_paradise_papers.2018-02-14/paradise_papers.nodes.address.csv', low_memory = False)
```

```python
edges.to_sql("edges",conn, if_exists = "replace")
address.to_sql("address", conn, if_exists = "replace")
```

```python
address.columns
```

```python
df = pd.read_sql("""select a.START_ID, b.countries from edges as a inner join 
(select START_ID, countries from edges inner join address on END_ID = node_id ) as b
on a.END_ID=b.START_ID
                """, conn)
```

```python
df
```

```python
df = df.rename(columns = {"START_ID":"node_id"})
df
```

```python
countries = df.append(pd.read_sql("""
select node_id, countries from address


""", conn), ignore_index = True)
countries
```

```python
edges.head()
```

```python
merged_df = edges.merge(countries, left_on="START_ID", right_on= "node_id")
```

```python
merged_df = merged_df.merge(countries, left_on="END_ID", right_on= "node_id", suffixes = ("_start", "_end"))
```

```python
filtered_df = merged_df[(merged_df["countries_start"] != merged_df["countries_end"]) & (merged_df["countries_start"].isnull() == False) & (merged_df["countries_end"].isnull() == False)]
```

```python
filtered_df.to_csv("filtered_countries.csv")
```
